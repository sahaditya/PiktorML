```python
# https://medium.com/dair-ai/implementing-a-logistic-regression-model-from-scratch-with-pytorch-24ea062cd856
# pytorch model from scratch

import torch
import torchvision
from torchvision import datasets, models, transforms

import os
import numpy as np
import matplotlib.pyplot as plt

print(f"torch version: {torch.__version__}")

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print(f"device: {device}")
```

```python
data_path = 'data/hymenoptera_data/'
```

```python
class ReshapeTransform:
    def __init__(self, new_size):
        self.new_size = new_size

    def __call__(self, img):
        result = torch.reshape(img, self.new_size)
        return result

## transformations used to standardize and normalize the datasets
data_transforms = {
    'train': transforms.Compose([
        transforms.Resize(224),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        ReshapeTransform((-1,)) # flattens the data
    ]),
    'val': transforms.Compose([
        transforms.Resize(224),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        ReshapeTransform((-1,)) # flattens the data
    ]),
}

## load the correspoding folders
image_datasets = {x: datasets.ImageFolder(os.path.join(data_path, x),
                                          data_transforms[x])
                  for x in ['train', 'val']}

## load the entire dataset; we are not using minibatches here
train_dataset = torch.utils.data.DataLoader(image_datasets['train'],
                                            batch_size=len(image_datasets['train']),
                                            shuffle=True)
test_dataset = torch.utils.data.DataLoader(image_datasets['val'],
                                           batch_size=len(image_datasets['val']),
                                           shuffle=True)
```

```python
## load the entire dataset
x, y = next(iter(train_dataset))

## print one example
dim = x.shape[1]
print("Dimension of image:", x.shape, "\n", 
      "Dimension of labels", y.shape)

plt.imshow(x[160].reshape(1, 3, 224, 224).squeeze().T.numpy())
```

```python
class LogisticRegression: 
    def __init__(self, dimension, lr=torch.scalar_tensor(0.01)):
        super(LogisticRegression, self).__init__()
        self.w = torch.zeros(dimension, 1, dtype=torch.float).to(device)
        self.b = torch.scalar_tensor(0).to(device)
        self.grads = {"dw": torch.zeros(dimension, 1, dtype=torch.float).to(device), "db": torch.scalar_tensor(0).to(device)}
        self.lr = lr.to(device)
    
    def forward(self, tensor):
        z = torch.mm(self.w.T, tensor)
        a = self.sigmoid(z)
        return a
    
    def sigmoid(self, z):
        s = 1/ (1 + torch.exp(-z))
        return s
    
    def backward(self, tensor, yHat, y):
        self.grads["dw"] = 1/tensor.shape[1] * torch.mm(tensor, (yHat-y).T)
        self.grads["db"] = 1/tensor.shape[1] * torch.sum(yHat-y)
    
    def optimize(self):
        self.w = self.w - self.lr * self.grads["dw"]
        self.b = self.b - self.lr * self.grads["db"]
        
def loss(yHat, y):
    m = y.size()[1]
    l = -(1/m) * torch.sum(y*torch.log(yHat) + (1-y)*torch.log(1-yHat))
    return l

def predict(yhat, y):
    y_prediction = torch.zeros(1, y.size()[1])
    for i in range(yhat.size()[1]):
        if yhat[0, i] <= 0.5:
            y_prediction[0, i] = 0
        else:
            y_prediction[0, i] = 1
    return 100 - torch.mean(torch.abs(y_prediction - y)) * 100
```

```python
x, y = next(iter(train_dataset))
print(f"x shape: {x.shape}")
x_flatten = x.T
print(f"x_flatten: {x_flatten.shape}")
y = y.unsqueeze(0) 
print(f"y shape: {y.shape}")
dim = x_flatten.shape[0]
print(f"dim: {dim}")
```

```python
model = LogisticRegression(dim)
yhat = model.forward(x_flatten.to(device))
yhat = yhat.data.cpu()
```

```python
cost = loss(yhat, y)
prediction = predict(yhat, y)
print("Cost: ", cost)
print("Accuracy: ", prediction)

model.backward(x_flatten.to(device), yhat.to(device), y.to(device))
model.optimize()
```

```python
#lets train model


## hyperparams
costs = []
dim = x_flatten.shape[0]
learning_rate = torch.scalar_tensor(0.0001).to(device)
num_iterations = 100
lrmodel = LogisticRegression(dim, learning_rate)


## transform the data
def transform_data(x, y):
    x_flatten = x.T
    y = y.unsqueeze(0) 
    return x_flatten, y 

## training the model
for i in range(num_iterations):
    x, y = next(iter(train_dataset))
    test_x, test_y = next(iter(test_dataset))
    x, y = transform_data(x, y)
    test_x, test_y = transform_data(test_x, test_y)

    # forward
    yhat = lrmodel.forward(x.to(device))
    cost = loss(yhat.data.cpu(), y)
    train_pred = predict(yhat, y)
        
    # backward
    lrmodel.backward(x.to(device), 
                    yhat.to(device), 
                    y.to(device))
    lrmodel.optimize()
    ## test
    yhat_test = lrmodel.forward(test_x.to(device))
    test_pred = predict(yhat_test, test_y)

    if i % 10 == 0:
        costs.append(cost)

    if i % 10 == 0:
        print("Cost after iteration {}: {} | Train Acc: {} | Test Acc: {}".format(i, 
                                                                                    cost, 
                                                                                    train_pred,
                                                                                    test_pred))
```
